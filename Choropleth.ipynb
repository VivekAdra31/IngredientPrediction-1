# Creating a Choropleth to show common ingredients by cuisine

### By finding most commonly occuring ingredients and manually choosing top 3 by eliminating secondary ingredients(such as salt and water)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
sb.set()
```

```python
#DataFrame manipulation
data = pd.read_json("data/train.json")
data = data.set_index("id")
data = data.sort_values("id")
data.head()
```

## Counting Ingredients In Every Cuisine:
Making use of nested dictionaries


```python
cuisine_list=[]
cuisine_list=data.cuisine.unique()
cuisine_dict = dict()
for i in cuisine_list:
    cuisine_dict[i]=dict()
```

```python
for i in range(0, 30000):
    for j in data.iloc[i]['ingredients']:
            if j in cuisine_dict[data.iloc[i]['cuisine']]:
                cuisine_dict[data.iloc[i]['cuisine']][j]+=1
            if j not in cuisine_dict[data.iloc[i]['cuisine']]:
                cuisine_dict[data.iloc[i]['cuisine']][j]=1
               
for i in cuisine_dict:
    print(i,": \n")
    print("{ \n")
    for j in cuisine_dict[i]:
        print(j,":",cuisine_dict[i][j])
    print("} \n")
    print("######################################################################################")
    print("######################################################################################")
            
```

## Modifying Data for the Choropleth:

```python
new_df1=pd.DataFrame(columns=['Ingredient', 'Code'])
for i in range(0,57):
    if i%3==0:
        data1={}
        data1['Ingredient']=[]
        data1['Code']=new_df.iloc[i]['Code']
        data1['Ingredient'].append(new_df.iloc[i]['Ingredient'])
    if i%3==1:
        data1['Ingredient'].append(new_df.iloc[i]['Ingredient'])
    if i%3==2:
        data1['Ingredient'].append(new_df.iloc[i]['Ingredient'])
        new_df1=new_df1.append(data1,ignore_index=True)
new_df1.head(n=30)

```
